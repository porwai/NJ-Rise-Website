```python
def format_column_name(col_name):
    formatted_name = ' '.join(word.capitalize() for word in col_name.replace('_', ' ').split())
    if formatted_name == "Client Id":
        return "Client Id (TEFAP)"
    if formatted_name == "Empoweror":
        return "EmpowerOR"
    return formatted_name
```

```python
# Assuming 'df' is your DataFrame
df.columns = [format_column_name(col) for col in df.columns]
```

```python
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import Font, Alignment
import pandas as pd

# Assuming 'df' is your DataFrame ready to export
wb = Workbook()
ws = wb.active

# Setup merged headers for "CITY/TOWN" and "HOUSING"
# CITY/TOWN from columns V to Y
ws.append([''] * 21 + ['CITY/TOWN', '', '', ''] + ['HOUSING', '', '', ''] + ['']*2 + ['Total return and new client', '', ''] + ['']*2)
ws.merge_cells(start_row=1, start_column=22, end_row=1, end_column=25)  # Columns V to Y for CITY/TOWN
ws.merge_cells(start_row=1, start_column=26, end_row=1, end_column=29)  # Columns Z to AC for HOUSING
ws.merge_cells(start_row=1, start_column=32, end_row=1, end_column=34)  # Columns AH to AH for New Client
ws.merge_cells(start_row=1, start_column=37, end_row=1, end_column=45)  # Columns AK to AS for Household Level Characteristics
ws.merge_cells(start_row=1, start_column=46, end_row=1, end_column=56)  # Columns 46 to 56 for Ethnicity
ws['V1'].alignment = ws['Z1'].alignment = Alignment(horizontal='center', vertical='center')
ws['V1'].font = ws['Z1'].font = Font(bold=True)

# Set the next row for data headers
header_row = 2

# Add DataFrame headers and format them
for col_idx, column_title in enumerate(df.columns, start=1):
    ws.cell(row=header_row, column=col_idx, value=column_title).font = Font(bold=True)
    ws.cell(row=header_row, column=col_idx).alignment = Alignment(horizontal='center')

# Add data rows starting from the next available row
data_start_row = header_row + 1
for r in dataframe_to_rows(df, index=False, header=False):
    ws.append(r)

# Add sum row immediately after the header row if the column is numerical
sum_row = header_row + 1
ws.insert_rows(sum_row)

for col_idx, col in enumerate(df.columns, start=1):
    if pd.api.types.is_numeric_dtype(df[col]):  # Check if the column is numeric
        # Calculate the sum and add it to the new row
        sum_value = df[col].sum()
        ws.cell(row=sum_row, column=col_idx, value=sum_value)

# Apply formatting to sum row
for cell in ws[sum_row]:
    cell.font = Font(bold=True)
    cell.number_format = '#,##0'  # Change format as necessary
    
#Set wrap text for all cells
for row in ws.iter_rows():
    for cell in row:
        cell.alignment = Alignment(wrap_text=True)
        
# Set the height of the first row
ws.row_dimensions[1].height = 30  # Adjust the height as needed

# Save the workbook
file_path = 'datatest/Formatted_MasterDB_Data2.xlsx'
wb.save(file_path)

print(f"File saved to {file_path}")
```
